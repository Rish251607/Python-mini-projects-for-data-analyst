```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("sales_data_2.csv", encoding="latin1")
df.info()
```

```python
df.isnull().sum()
```

```python
df['STATE'] = df['STATE'].fillna('notdefined', inplace=True)
df.isnull().sum()
```

```python
df['TERRITORY'] = df['TERRITORY'].fillna('notdefined', inplace=True)
df['POSTALCODE'] = df['POSTALCODE'].fillna('notdefined', inplace=True)
df.isnull().sum()
```

```python
df['ADDRESSLINE2'] = df['ADDRESSLINE2'].fillna('notdefined')
df.isnull().sum()
```

```python
df['ORDERDATE'] = pd.to_datetime(df['ORDERDATE'])
df['Month'] = df['ORDERDATE'].dt.month
df['Year'] = df['ORDERDATE'].dt.year
```

```python
df.info
```

```python
df.info()
```

```python
monthly_sale = df.groupby(['Year','Month'])['SALES'].sum()
monthly_sale
```

```python
monthly_sale.plot(kind='bar')
plt.title('Monthly sales')
plt.xlabel('month')
plt.ylabel('revenue')
plt.show()
```
